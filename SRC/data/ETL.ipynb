```python
import pandas as pd
import numpy as np

df = pd.read_excel("../../data//raw/Ajustes_inventarios.xlsx")
df.columns = df.columns.str.strip() # eliminacion de espacios en blanco en los nombres de las columnas
df.columns = df.columns.str.replace(".", "_") # reemplazo de espacios en blanco por guion bajo

df = df.rename(columns={
    "Ctd_diferencia": "stock_diferencia",
    "Impte_diferenc_": "impacto_valor",
    "FeRecuento": "fecha",
    "Cantidad teórica": "cantidad_teorica",
    "Cantidad contada": "cantidad_contada",
    "Contado por": "usuario",
    "Ce_": "centro",
    "Alm_": "almacen",
    "Material": "material",
    "Tipo stocks": "tipo_stock"
    
    
})

df.head()
```

observamos que hay columnas que no seran necesaria para el analisis, como es el caso de las columnas: [ValDiferInvent posit]; [Valor difer.negativo]; [Pos.]; [Ce.]

Adicionalmente se observa que la columna [FeRecuento] que es la que almacena la fecha de realizacion de los inventarios esta como tipo string, lo cual tendremos que transformar al formato fecha

```python
# Cambio de formato a fecha 
df["fecha"] = pd.to_datetime(df["fecha"], format="%d.%m.%Y")
```

Teniendo ya la fecha en el formato correcto, procederemos a limpiar la tabla de datos innecesarios como lo es las filas con diferencia de 0

```python
#eliminar filas sin diferencia de inventario
df = df[df["stock_diferencia"] != 0]

#eliminar ajustes de usuarios diferentes al pretoria
df = df[df["usuario"] == "PP1300_SOAP"]

#seleccionar columnas relevantes
df = df[["fecha", "centro", "almacen", "tipo_stock","material", "cantidad_teorica", "cantidad_contada", "stock_diferencia", "impacto_valor"]]

df.info()
```
